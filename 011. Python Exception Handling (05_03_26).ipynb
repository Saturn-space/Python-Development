```python

try:
  num1 = int(input("Enter a number : "))
  num2 = int(input("Enter a number : "))
  result = num1/num2;

  print(result)

except ValueError:
  print("Please Enter the values correctly")

except ZeroDivisionError:
  print("Division By zero not allowed")

except TypeError:        # only used when dealing with multiple types of data like adding charater and int value together.
  print("Wrong type of values")

finally:
  print("Program Executed Successfully !")

```

```python
try:
  val1 = int(input("Enter a value : "))
  val2 = input("Enter a value : ")

  result = val1 + val2
  print("Result :",result)

except ValueError:
  print("Wrong values")

except TypeError:
  print("Different type of value")

finally:
  print("Program Completed")
```
