```python
#Playing with Pandas:

import pandas as pd
data = {
    "Maths":[22,44,66],
    "Science":[72,66,26],
    "English":[62,54,16]
}

df = pd.DataFrame(data)

#index
dfi = pd.DataFrame(data,index=["Rohan","Sohan","Dohan"])
print("\n",df)
print("\n",dfi)
print("\n",df[["English", "Maths"]])

#loc
print(df.loc[2,"Maths"])
print(df.loc[:,:])
print(dfi.loc[:,:])
print(df.loc[:,"Maths"])

#iloc
print(df.iloc[2,2])
print(df.iloc[0,2])
print(df.iloc[2,0])
```

```python
#1. Write a python program using pandas dataframe:
#a. Create a dataframe of 5 students and 3 subjects

marks = {
    "Maths":[45,23,76,43,88],
    "Physics":[33,77,12,89,34],
    "Hindi":[34,67,45,90,65]
}
name = ["Nohan","Rohan","Sohan","Kohan","Dohan"]
dfi = pd.DataFrame(marks,name)

#b.Display the dataframe
print(dfi)
print("\n",dfi.loc[:,:])

#c.Find Total and average marks of each student
dfi["Total"] = dfi.sum(axis=1)
dfi["Average"] = dfi[["Maths","Physics","Hindi"]].mean(axis = 1)
print("\n",dfi.loc[:,:])

#d. Highest Marks in each subject
print("\nHighest Marks in each Subject : \nMaths : ",dfi["Maths"].max(axis=0),"\nPhysics : ",dfi["Physics"].max(axis=0),"\nHindi : ",dfi["Hindi"].max(axis=0))
```
